# Guided Capstone Project Report - Big Mountain Ski Resort
## Context and Business Problem
The Big Mountain ski resort in Montana receives around 350,000 people every year for skiing and snowboarding activities. So far, the company has used a market-oriented pricing strategy, which consists of charging a premium over the market price average. However, our client is considering a different pricing approach and is looking to increase its revenue. Revenue increase might be accomplished in one of two ways: (1) by increasing ticket prices even higher or (2) by cutting operational costs and therefore increasing profit margins. In this report, I focus on strategy (1) given that the data available for this analysis does not include enough information on costs.

## Target Feature and Market Size
To achieve the clientâ€™s goal of improving revenue, I focus my analysis on ticket prices and how the resort might be able to charge a higher premium. While the available data contained both weekend and weekday ticket prices, I decided to focus only on weekend prices for three reasons: (1) the available dataset contained more missing values for weekday prices than weekend prices, (2) there is a roughly 1-to-1 correlation between weekend and weekday ticket prices once the value was above \$100 (Fig. 1A), and (3) our resort currently charges the same ticket prices for both weekends and weekdays. Further, Principal Component Analysis (PCA) of this price data, which is able to explain over 77\% of price variability, suggests no obvious pricing pattern across states (Fig. 1B). For this reason, the analysis presented assumes a nation-wide ski resort market.

## Main Features Controlling Market Prices
Modeling results suggest 8 features as the main drivers of price, namely vertical drop, snow making acreage, total number of chairs, number of fast Quads, number of runs, length of longest run, number of trams, and the size of the skiable terrain. Big Mountain has a solid standing in the market for most of these features, as can be seen in Figure 2. ## Recommendation
Scenarios 2 and 3 are the only ones that would yield further increases in ticket price. However, scenario 3 is likely to result in higher operational costs than scenario 2, given the need for more snow making machines and machine maintenance. The ROI of scenario 3 is not that much bigger than scenario 2, so I would suggest scenario 2 to business leadership. If possible, our client could do both 1 and 2: close one run (with no penalty on price) and open another one that would allow for the increase in vertical drop. Finally, the additional chair will cost the business \\$1,540,000 in operational cost, which would amount to \\$0.88 per ticket going into cost coverage. All else equal, this would result in a net revenue of \\$13,525,471 (for the combined scenarios 1 and 2).
